```python
import pandas as pd
```

```python
# Cargar el archivo de Excel en un dataframe
df = pd.read_excel('/content/HepatitisCdata.xlsx')
df
```

```python
# Ver las primeras filas del dataframe
print(df.head())
```

```python
# Eliminar los datos NaN del dataframe
df = df.dropna()

# Mostrar el dataframe sin los datos NaN
print('Dataframe sin los datos NaN:')
print(df)
```

```python
df = df.drop(['patient number', 'Age', 'Sex'], axis=1)
```

```python
df
```

```python
import sklearn 
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder() 
df['Donor Blood'] = lb.fit_transform(df['Donor Blood'])
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


top = 15
corr = df.corr()
top15 = corr.nlargest(top, 'Donor Blood')['Donor Blood'].index
corr_top15 = df[top15].corr()
f,ax = plt.subplots(figsize=(10,10))
sns.heatmap(corr_top15, square=True, ax=ax, annot=True, cmap='coolwarm', fmt='.2f', annot_kws={'size':12})
plt.title('Top correlated features of dataset', size=16)
plt.show()
```

```python
x = df.drop(['Donor Blood'], axis=1)
x     
```

```python
y = df.loc[:,'Donor Blood']
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.15, random_state=42)
```

```python
model = LogisticRegression()
model
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
decision_tree = DecisionTreeClassifier(max_depth=5) 
```

```python
decision_tree.fit(x_train, y_train) 
```

```python
y_pred = decision_tree.predict(x_test)
```

```python
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python
print(classification_report(y_test,y_pred))
```
